```python
diet_ingredients = pd.DataFrame(replace_data[['diet', 'ingredients']])
diet_ingredients['ingredients'] = diet_ingredients['ingredients'].str.split(',')
diet_ingredients = diet_ingredients.explode('ingredients')
```

```python
diet_ingredients
```

```python
# Group by the 'Diet' column and count the occurrences of each ingredient
diet_ingredients_count = diet_ingredients.groupby('Diet')['ingredients'].value_counts().reset_index(name='count')
diet_ingredients_count.head(20)
```

```python
top_5_ingredients = diet_ingredients_count.groupby('Diet').apply(lambda x: x.nlargest(5,'count')).reset_index(drop=True)
top_5_ingredients
```
