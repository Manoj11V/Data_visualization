```python
import pandas as pd
data=pd.read_csv('/content/drive/MyDrive/HVA WORKS/Ifood_new.csv')
replace_data=pd.read_csv('/content/drive/MyDrive/HVA WORKS/Ifood_new.csv')
data
```

```python
replace_data.replace(-1,pd.NA,inplace=True)
replace_data.replace('-1',pd.NA,inplace=True)
replace_data
```

```python
replace_data.dropna(inplace=True)
replace_data=pd.DataFrame(replace_data)
replace_data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
average_prep_time=replace_data.groupby('diet')['prep_time'].mean().reset_index()
lineplot=sns.lineplot(data=average_prep_time,x=replace_data['diet'],y=replace_data['prep_time'], markers=True, dashes=False,palette='tab10')
plt.title('Average Prep Time by Diet')
plt.xlabel('Diet')
plt.ylabel('Average Prep Time (minutes)')
plt.show()
```

```python
line_plot_matplotlib=plt.plot(average_prep_time['diet'],average_prep_time['prep_time'])
plt.title('Average Prep Time by Diet')
plt.xlabel('Diet')
plt.ylabel('Average Prep Time (minutes)')
plt.show
```

```python
average_cook_time=replace_data.groupby('diet')['cook_time'].mean().reset_index()
average_cook_time=pd.DataFrame(average_cook_time)
average_prep_time=pd.DataFrame(average_prep_time)
print((average_prep_time))
average_cook_time
```

```python
avg_times = replace_data.groupby(['diet']).agg({
    'prep_time': 'mean',
    'cook_time': 'mean'
}).reset_index()
for diet in avg_times['diet']:
    plt.plot(average_cook_time['cook_time'],average_prep_time['prep_time'], label=diet)
plt.show()
```

```python
dish_counts = replace_data['diet'].value_counts()
```

```python
dish_counts.plot(kind='bar', color=['green', 'red'])
plt.title('Count of Dishes by Diet Type')
plt.xlabel('Diet Type')
plt.ylabel('Number of Dishes')
plt.show()
```

```python
region_counts = replace_data['region'].value_counts()
```

```python
region_counts.plot(kind='barh', color='skyblue')
plt.title('Total Number of Dishes by Region')
plt.xlabel('Number of Dishes')
plt.ylabel('Region')
plt.show()
```

```python
region_diet_counts = replace_data.groupby(['region', 'diet']).size().unstack(fill_value=0)
region_diet_counts.plot(kind='bar', stacked=True, color=['green', 'red'])
plt.title('Distribution of Diet Types by Region')
plt.xlabel('Region')
plt.ylabel('Count of Dishes')
plt.legend(title='Diet Type')
plt.show()
```

```python
plt.hist(replace_data['prep_time'], bins=10, edgecolor='black', color='skyblue')
plt.title('Distribution of Preparation Times')
plt.xlabel('Preparation Time (minutes)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(replace_data['prep_time'],bins=5,edgecolor='blue',color='red')
plt.title('Distribution of Preparation Times')
plt.xlabel('Preparation Time (minutes)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.scatter(replace_data['prep_time'], replace_data['cook_time'], color='purple')
plt.title('Preparation Time vs Cooking Time')
plt.xlabel('Preparation Time (minutes)')
plt.ylabel('Cooking Time (minutes)')
plt.show()
```

```python
diet_counts = replace_data['diet'].value_counts()
plt.pie(diet_counts, labels=diet_counts.index, autopct='%1.1f%%', startangle=140, colors=['green', 'red'])
plt.title('Diet Type Distribution')
plt.show()
```

```python
plt.pie(diet_counts, labels=diet_counts.index, autopct='%1.1f%%', startangle=140, colors=['green', 'red'], wedgeprops=dict(width=0.3))
plt.title('Diet Type Distribution')
plt.show()
```
